```python
#The boxplot() method in Pandas is the standard way of showing the distribution of data through their quartiles.
#A box plot displays the distibution of the data based ;on the foive-number summary:
#Minimum, first quartile(Q11), median, third quartile(Q3), and the maximum.

#Box Components:
#The box spans the interquartile range(IQR), representaing the middle 50% of the data.
#Median(): A line inside the box marking the 50th percentile.
#Whiskers: Lines extending from the box to the minimum and the maximum non-outlier values.
#Outliers: Individual points plotted beyond the whiskers
#Median (Q2/50th Percentile): Middle line in the box.
#First Quartile(Q1/ 25th Percentile): Bottom of the box.
#Third Quartile(Q3/ 75th Percentile): Top of the box.
#Whiskers: Extended from the box, to the min/max data point.
```
